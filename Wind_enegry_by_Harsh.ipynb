```python
X=train.drop(['Target','timestamp'],axis=1)
y=train['Target']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.decomposition import PCA
```

```python
p=PCA(n_components=0.95)
pca_data=p.fit_transform(X)
pca_data.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(pca_data,y,test_size=0.3,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
```

```python
from sklearn.metrics import *
```

```python
r2_score(y_test,prediction)
```

```python
#decision tree regressor
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(X_train,y_train)
```

```python
predictions_dt=dt.predict(X_test)
```

```python
r2_score(y_test,predictions_dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
predictions_rf=rf.predict(X_test)
```

```python
r2_score(y_test,predictions_rf)
```
